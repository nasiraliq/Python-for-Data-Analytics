```python
import pandas as pd
```

```python
df = pd.read_csv("data/data_jobs.csv")
```

```python
df["job_posted_date"] = pd.to_datetime(df["job_posted_date"])
```

```python
df.info()
```

```python
df_us = df[df["job_country"].isin(["United States"])]
```

```python
df.dropna(subset=["salary_year_avg"], inplace=True)
```

```python
df.info()
```

```python
df_us.groupby(by=["job_title_short"])["salary_year_avg"].agg(["median", "min", "max", "count"]).sort_values(by=["median", "count"], ascending=False).plot(figsize=(14,6))
```
