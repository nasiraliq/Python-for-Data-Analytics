```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# import seaborn as sns
```

```python
df = pd.read_csv("data/data_jobs.csv")
```

```python
df["job_posted_date"] = pd.to_datetime(df["job_posted_date"])
```

```python
df_new = df.dropna(subset=["salary_year_avg"])
```

```python
plot = plt.hist(x=df_new["salary_year_avg"], bins=30, color="skyblue", edgecolor="black", alpha=0.7)
plt.show()
```
