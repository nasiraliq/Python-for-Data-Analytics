```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data/data_jobs.csv")
df["job_posted_date"] = pd.to_datetime(df["job_posted_date"])
df.sort_values(by="job_posted_date", inplace=True, ascending=True)
```

```python
df["job_posted_month"] = df["job_posted_date"].dt.month
```

```python
job_posted_month_count = df.job_posted_month.value_counts().sort_index()
```

```python
plt.figure(figsize=(16,5))
plt.plot(job_posted_month_count.index, job_posted_month_count.values)
plt.show()
```

```python
jobs_posted_counts = df["job_title_short"].value_counts().sort_values(ascending=False)
```

```python
plt.figure(figsize=(16,5))
plt.bar(jobs_posted_counts.index, jobs_posted_counts.values)
plt.xticks(rotation=45, ha="right")
plt.title("Postings by Job Title")
plt.ylabel("Count of Job Postings")
plt.show()
```

```python
job_schedule_type_count = df["job_schedule_type"].value_counts().sort_values(ascending=False).iloc[1:11]
plt.bar(job_schedule_type_count.index, job_schedule_type_count.values, align="center", width=0.9)
plt.title("Job Schedule Type")
plt.ylabel("Count of Job Schedule Type")
plt.xticks(rotation=45, ha="right")
plt.show()
```
